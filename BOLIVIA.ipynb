```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")


driver_path = ChromeDriverManager().install()


driver = webdriver.Chrome(driver_path, chrome_options=options)
                     
                     
# Inicializamos el navegador
driver.get('https://www.segip.gob.bo/inicio/bolivianos/')
sleep(1)

#bajamos con scroll
driver.execute_script("window.scrollTo(0, 2500)")

j = 3
box3 = 'xxx'
while box3 != 1:
    driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{j}]/div[1]/i').click()
    sleep(1)
    try:
        driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{j+1}]/div[1]/i')
    except:
        box3 = 1
    j = j + 1



depart =  []
prov   =  []
muni   =  []
direcc =  []
serv   =  []
horar  =  []

### 1ERO EXTRAEMOS INFORMACION DE LA PAZ Y COCHABAMBA:
x = 3
box2 = "xx"    
while box2 != 20000:
    if x > 6:
        break
    
    i = 2
    box = "x"    
    while box != 0:
        
        departamento =  [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i}]/td[1]').text]

        provincias   =  [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i}]/td[2]').text]

        municipios   =  ['NaN']
        
        direcciones  =  [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i}]/td[3]').text]

        servicios    =  [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i}]/td[4]').text]

        horarios     =  [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i}]/td[5]').text]

        depart =  depart + departamento
        prov   =   prov  + provincias
        muni   =  muni   + municipios
        direcc =  direcc + direcciones
        serv   =  serv   + servicios
        horar  =  horar  + horarios
        try:
            driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x}]/div[2]/table/tbody/tr[{i+1}]/td[1]').text
        except:
            box = 0

        i = i + 1
    try:
        driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{x+3}]/div[2]/table/tbody/tr[2]/td[1]').text
      
    except:
        
        box2 = 20000

    x = x + 3 
    
    
### AHORA EXTRAEMOS INFORMACION DE LOS 6 DEPARTAMENTO RESTANTES:
sleep(1)

p = 4
autos = "p"    
while autos != 40000:
    
    if p == 6:
        p = p+1
    if p == 9:
        p = p+1
    
    departamento =   [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[1]/h4').text]

    o = 2
    auto = "o"    
    while auto != 30000:


        provincias   =   ['NaN']

        municipios   =   [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[2]/table/tbody/tr[{o}]/td[1]').text]

        direcciones  =   [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[2]/table/tbody/tr[{o}]/td[2]').text]

        horarios     =   [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[2]/table/tbody/tr[{o}]/td[3]').text]

        servicios    =   [driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[2]/table/tbody/tr[{o}]/td[4]').text]

        depart =  depart + departamento
        prov   =   prov  + provincias
        muni   =  muni   + municipios
        direcc =  direcc + direcciones
        horar  =  horar  + horarios
        serv   =  serv   + servicios


        try:
            driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p}]/div[2]/table/tbody/tr[{o+1}]/td[1]').text

        except:

            auto = 30000

        o = o + 1
        
    try:
        driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/main/section/div/article/div[2]/div/div/div/div[{p +1}]/div[1]/h4').text
      
    except:
        
        autos = 40000

    p = p + 1 
    

# SANTA CRUZ TIENE DIFERENCIAS DE PATRON PARA INCLUIRLO EN EL LOOP:
# SE ENCONTRARON PROBLEMAS CON LA SECUENCIA DEL HTML, UN LOOP CON CORRECTA NUMERACIÓN 
# JALABA DATOS DE DIFERENTES COLUMNAS 
# POR ESA RAZÓN SE ELABORO "SANTA CRUZ" A DEDO 
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
   



horizontal = driver.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[2]/table/tbody/tr')

for tr in horizontal:
    
    departamm = tr.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[1]/h4')
    pro       = ['NaN']
    municipio = tr.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[2]/table/tbody/tr/td[1]')
    direccion = tr.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[2]/table/tbody/tr/td[2]')
    horario   = tr.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[2]/table/tbody/tr/td[3]')
    servicios = tr.find_elements_by_xpath('//*[@id="1470170671047-f753fac3-0e45"]/div[2]/table/tbody/tr/td[4]')

    
    
dep = list( map(lambda x: x.text, departamm  ) )
    
    
    
#Limpiamos información innecesaria    
horarii = list( map(lambda x: x.text, horario  ) )
del horarii[0]    
    
municip = list( map(lambda x: x.text, municipio) )

#Limpiamos elementos vacios e información innecesaria
del municip[0]
del municip[5]
del municip[5]

#Colocamos los horarios de "municip" en su lugar "horarii"
if horarii != 0:
    horarii[1] = horarii[1] + ('\n') + municip[2] + ('\n') + municip[3]
    
#podemos limpiar los datos mal ubicados en "municip"
del municip[2]
del municip[2]


#Limpiamos elementos vacios e información innecesaria
direccc = list( map(lambda x: x.text, direccion) )
del direccc[0]
del direccc[2]
del direccc[2]


#Limpiamos información innecesaria
servicc = list( map(lambda x: x.text, servicios) )
del servicc[0]

depart =  depart + dep  * len(municip)
prov   =   prov  + pro  * len(municip)
muni   =  muni   + municip
direcc =  direcc + direccc
serv   =  serv   + servicc
horar  =  horar  + horarii

# PONDREMOS TODO "depart" EN MAYUSCULA

depa = list( map(lambda x: x.upper(), depart) )


    

                                              
```

```python
pd.DataFrame({
    'DEPARTMENT' : depa,
    'PROVINCE' : prov,
    'MUNICIPALITY' : muni,
    'ADDRESS': direcc,
    'SERVICE': serv,
    'HOURS OF ATTENTION' : horar
}).to_excel(r"BOLIVIA.xlsx" , index = False )
```
